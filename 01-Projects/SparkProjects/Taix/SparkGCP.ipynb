```python
df=spark.read.parquet('gs://mydata0101/yellow_tripdata_2022-01.parquet')
```

```python
df.count()
```

```python
%load_ext google.cloud.bigquery
```

```python
from pyspark.sql.functions import *
df = df.withColumn("date_pickup", to_date("tpep_pickup_datetime"))
df = df.withColumn("time_pickup", date_format(
    col("tpep_pickup_datetime"), "HH:mm:ss"))
df = df.withColumn("date_dropoff", to_date("tpep_dropoff_datetime"))
df = df.withColumn("time_dropoff", date_format(
    col("tpep_dropoff_datetime"), "HH:mm:ss"))
cols = (["tpep_pickup_datetime", "tpep_dropoff_datetime"])
df = df.drop(*cols)
df.printSchema()
# df.show()
```

```python
df.repartition(1).write.parquet(
    'gs://mydata0101/data')
```
