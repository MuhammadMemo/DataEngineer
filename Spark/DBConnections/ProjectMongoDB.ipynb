mongodb://localhost:27017/

```python
import pymongo
myclient = pymongo.MongoClient("mongodb://localhost:27017/")
mydb = myclient["Mydb"]
```

fetch Customer data

```python
Query_CustomerID= {'CustomerID': { "$in" : [ 75857,55025 ] } }
CustomerColl = mydb["Customer"].find(Query_CustomerID)#.limit(100)
```

fetch order Data

```python
Query_Order= {'CustomerID': { "$in" : [ 75857,55025 ] } }
OrderCall=mydb["Order"].find(Query_Order)#.limit(100)
```

Convert data to dataframe

```python
import pandas as pd
```

```python
df_Customer=pd.DataFrame(list(CustomerColl))
df_Customer
```

```python
df_Customer.columns
```

```python
df_order= pd.DataFrame(list(OrderCall))
df_order
```

```python
df_order.columns
```

```python
df_order.describe()
```

```python
df_Customer=df_Customer.set_index('CustomerID')
df_order=df_order.set_index('CustomerID')
```

```python
customer_order=df_Customer.\
    join(df_order,how='left',  lsuffix='_Customer')
customer_order
```

```python
customer_order.shape
```

groupby Orders by customers

```python
customer_order.reset_index('CustomerID',inplace=True)
```

```python
sumOrder=customer_order.groupby('CustomerID')['AmountTax'].\
    agg(sumOrder='sum').\
    reset_index().\
    query('sumOrder <=80000 ')
    #query('sumOrder in (1,3) ')
```

```python
sumOrder
```

other groupby Orders by customers

```python
customer_order.groupby("CustomerID", group_keys=False).sum('AmountTax').reset_index()
```
